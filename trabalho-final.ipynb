# RMS Titanic

O RMS Titanic foi um navio de passageiros britânico operado pela White Star Line e construído pelos estaleiros da Harland and Wolff em Belfast. Foi a segunda embarcação da Classe Olympic de transatlânticos depois do RMS Olympic e seguido pelo HMHS Britannic. Projetado pelos engenheiros navais Alexander Carlisle e Thomas Andrews, sua construção começou em março de 1909 e ele foi lançado ao mar em maio de 1911. O Titanic foi pensado para ser o navio mais luxuoso e mais seguro de sua época, gerando lendas que era supostamente "inafundável".

A embarcação partiu em sua viagem inaugural de Southampton para Nova Iorque em 10 de abril de 1912, no caminho passando em Cherbourg-Octeville na França e por Queenstown na Irlanda. Ele colidiu com um iceberg às 23h40min do dia 14 de abril e afundou na madrugada do dia seguinte matando mais de 1500 pessoas, sendo um dos maiores desastres marítimos em tempos de paz de toda a história. Seu naufrágio destacou vários pontos fracos de seu projeto, deficiências nos procedimentos de evacuação de emergência e falhas nas regulamentações marítimas da época. Comissões de inquérito foram instauradas nos Estados Unidos e no Reino Unido, levando a mudanças nas leis internacionais de navegação que permanecem em vigor mais de um século depois.

Os destroços do Titanic foram procurados por décadas até serem encontrados em 1985 por uma equipe liderada por Robert Ballard. Ele se encontra a 3843 m de profundidade e a 650 km ao sudeste de Terra Nova no Canadá. Sua história e naufrágio permaneceram no imaginário popular durante décadas, levando a produção de vários livros e filmes a seu respeito, mais notavelmente o filme Titanic de 1997. Até hoje o Titanic permanece como um dos navios mais famosos da história, com seus destroços atraindo várias expedições de exploração ao longo dos anos.

## Adequando o DataSet

5 - Uma vez que algumas colunas não serão utilizadas, eventualmente é melhor excluí-las para que não interfiram na análise. Assim, exclua do dataset as colunas Sibsp, Parch e Ticket.

6 - Renomear as colunas restantes para a lingua portuguesa, utilizando os seguintes nomes de colunas: IdPassageiro, Sobreviveu, Classe, Nome, Sexo, Idade, Tarifa, Cabine e Embarque.

7 - Alterar o conteudo da coluna Sobreviveu para:
- 0 => Não
- 1 => Sim

8 - Alterar o conteudo da coluna Sexo para:
- female => Mulher
- male   => Homem

```python
class Adequar:
  def __init__(self, df):
    self.__df = df

  @property
  def df(self):
    return self.__df

  def preprocess_data(self):
    self._drop_columns()
    self._rename_columns()
    self._map_values()
    self._fill_missing_values()
    return self.__df

  def _drop_columns(self):
    columns_to_drop = ['SibSp', 'Parch', 'Ticket']
    self.__df = self.__df.drop(columns=columns_to_drop)

  def _rename_columns(self):
    column_mapping = {
        "PassengerId": "IdPassageiro",
        "Survived": "Sobreviveu",
        "Pclass": "Classe",
        "Name": "Nome",
        "Sex": "Sexo",
        "Age": "Idade",
        "Fare": "Tarifa",
        "Cabin": "Cabine",
        "Embarked": "Embarque",
    }
    self.__df.rename(columns=column_mapping, inplace=True)

  def _map_values(self):
    self.__df['Sobreviveu'] = self.__df['Sobreviveu'].map({1: 'Sim', 0: 'Não'})
    self.__df['Sexo'] = self.__df['Sexo'].map({'male': 'Homem', 'female': 'Mulher'})

  def _fill_missing_values(self):
    if self.__df['Idade'].isnull().any():
      mean_age = round(self.__df['Idade'].mean())
      self.__df['Idade'].fillna(mean_age, inplace=True)
      self.__df['Idade'] = self.__df['Idade'].astype(int)

    if self.__df['Embarque'].isnull().any():
      mode_value = self.__df['Embarque'].mode()[0]
      self.__df['Embarque'].fillna(mode_value, inplace=True)
```

```python
adequar = Adequar(tit)
tit = adequar.preprocess_data()
tit
```

###Classe Passageiros

```python
class Passageiros:
  def __init__(self, dataset):
    self.__df = dataset

  @property
  def df(self):
    return self.__df

  def contar_sexos(self):
    sexo_passageiros = self.__df['Sexo'].value_counts()
    print(f'A análise mostra que temos {sexo_passageiros["Homem"]} passageiros homens e {sexo_passageiros["Mulher"]} mulheres.')

  def grafico_sobreviventes(self):
    return self.__df['Sobreviveu'].value_counts().plot.pie(colors=('tab:red', 'tab:blue'),
                                                    title='Percentual de sobreviventes',
                                                    fontsize=15,
                                                    shadow=True,
                                                    startangle=90,
                                                    autopct='%1.1f%%',
                                                    labels=('Não sobreviveram', 'Sobreviveram'),
                                                    figsize=(6,4)).set_ylabel('')
  def contar_sobreviventes(self):
    dados_sobreviventes = self.__df['Sobreviveu'].value_counts()
    print(f'A análise mostra que sobreviveram {dados_sobreviventes["Sim"]} passageiros. E {dados_sobreviventes["Não"]} passageiros não sobreviveram')

  def contar_mulheres_nao_sobreviventes(self):
    mulheres_nao_sobreviventes = self.__df[(self.__df['Sexo'] == 'Mulher') & (self.__df['Sobreviveu'] == 'Não')]
    total = mulheres_nao_sobreviventes[mulheres_nao_sobreviventes.columns[0]].count()
    return print(f'A análise mostra que {total} mulheres não sobreviveram.')

  def proporcao_sobrevivencia_sexo(self):
    sobreviveu = self.__df['Sobreviveu'] == 'Sim'
    sobreviventes_por_sexo = self.__df[sobreviveu]['Sexo'].value_counts()

    return sobreviventes_por_sexo.plot.pie(colors=('tab:red', 'tab:blue'),
                                    title='Percentual de sobreviventes por sexo',
                                    fontsize=15,
                                    shadow=True,
                                    startangle=90,
                                    autopct='%1.1f%%',
                                    labels=('Mulheres', 'Homens'),
                                    figsize=(6,4)).set_ylabel('')

  def idade_maior_numero_mortos(self):
    nao_sobreviveu = self.__df[self.__df['Sobreviveu'] == 'Não']

    idade_maior_mortos = nao_sobreviveu['Idade'].value_counts().idxmax()
    quantidade_mortos = nao_sobreviveu['Idade'].value_counts().max()
    return print(f'A análise mostra que passageiros na faixa dos {idade_maior_mortos} anos foram os que menos sobreviveram, tendo {quantidade_mortos} mortos.')


  def grafico_numero_mortos_por_idade(self):
    nao_sobreviveu = self.__df[self.__df['Sobreviveu'] == 'Não']

    mortos_por_idade = nao_sobreviveu['Idade'].value_counts().sort_index()

    plt.figure(figsize=(10, 6))
    plt.bar(mortos_por_idade.index, mortos_por_idade.values, color='red')
    plt.title('Número de não sobreviventes por idade')
    plt.xlabel('Idade')
    plt.ylabel('Número de Mortos')
    plt.show()

  def media_idade_homens_sobreviventes(self):
    homens_sobreviventes = self.__df[(self.__df['Sexo'] == 'Homem') & (self.__df['Sobreviveu'] == 'Sim')]
    media = round(homens_sobreviventes['Idade'].mean())
    return print(f'A análise mostra que a média de idade dos homens sobreviventes é de {media} anos.')

  def proporcao_sobrevivencia_prioritarios_sexo(self):
    passageiros_prioritarios = self.__df[(self.__df['Sexo'] == 'Mulher') | ((self.__df['Idade'] <= 15) & (self.__df['Idade'] >= 0))]

    sobreviveu = passageiros_prioritarios['Sobreviveu'] == 'Sim'
    sobreviventes_por_sexo = passageiros_prioritarios[sobreviveu]['Sexo'].value_counts()

    return sobreviventes_por_sexo.plot.pie(colors=('tab:red', 'tab:blue'),
                                        title='Proporção de sobreviventes entre passageiros prioritários por sexo',
                                        fontsize=15,
                                        shadow=True,
                                        startangle=90,
                                        autopct='%1.1f%%',
                                        labels=('Mulheres', 'Homens'),
                                        figsize=(6, 4)).set_ylabel('')


  def quantidade_passageiros_por_classe(self):
    classe =  self.__df['Classe'].value_counts()
    classe_um = classe[1]
    classe_dois = classe[2]
    classe_tres = classe[3]

    return print(f'A análise mostra que temos: \
    \n- {classe_um} passageiros na primeira classe. \keep
    \n- {classe_dois} passageiros na segunda classe. \
    \n- {classe_tres} passageiros na terceira classe.')

  def percentual_sobreviventes_por_classe(self):
    mapeamento_classe = {1: '1º Classe', 2: '2º Classe', 3: '3º Classe'}
    sobreviveu = self.__df['Sobreviveu'] == 'Sim'
    sobreviventes_por_classe = self.__df[sobreviveu]['Classe'].map(mapeamento_classe).value_counts()

    return sobreviventes_por_classe.plot.pie(colors=('#008000', '#0000FF', '#FFFF00'),
                                    title='Percentual de sobreviventes por Classe',
                                    fontsize=15,
                                    shadow=True,
                                    startangle=90,
                                    autopct='%1.1f%%',
                                    labels=('1º Classe', '3º Classe', '2º Classe'),
                                    figsize=(6,4)).set_ylabel('')

  def criar_dataframe_sobreviventes_por_sexo_classe(self):
    df_sobreviventes = (
        self.__df.groupby(['Sexo', 'Classe', 'Sobreviveu'])
        .size()
        .unstack(fill_value=0)
        .astype(int)
        .rename(columns={0: 'Não Sobreviveu', 1: 'Sobreviveu'})
    )

    return df_sobreviventes

  def sobreviventes_24_30(self):
    homens_classe_3_sobreviveram = self.__df.query("Sexo == 'Homem' and Classe == 3 and Sobreviveu == 'Sim' and Idade.between(24, 30)").shape[0]
    homens_classe_2_nao_sobreviveram = self.__df.query("Sexo == 'Homem' and Classe == 2 and Sobreviveu == 'Não' and Idade.between(24, 30)").shape[0]

    return print(f"Homens entre 24 e 30 anos da 3º classe que sobreviveram: {homens_classe_3_sobreviveram}\
    \nHomens entre 24 e 30 anos da 2º classe que não sobreviveram: {homens_classe_2_nao_sobreviveram}")


passageiros = Passageiros(tit)
```
